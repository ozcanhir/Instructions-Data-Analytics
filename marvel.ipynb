```python
#IMport Data and read it 
import pandas as pd 
marvel= pd.read_csv(r"C:\Users\Pc\Desktop\Data-Analytics-BI\Dataset\imdb_com_movie_data_promotion_promptcloud_master_deduped_n-20190529_123020337730168__0 2.csv")
print(marvel.head(10))
```

```python
# Data Info 
marvel.info()
```

```python
# Check Nan
print(marvel.isna().sum())
```

```python
# Dropp NaN in Column Budget
print(marvel["Budget"].isna())

```

```python
#Drop NaN from Budget
marvelV0=marvel.drop([0,6,10])
marvelV0.isna().sum()
```

```python
marvelV0['Collection'] = marvelV0['Collection'].str.replace('\W', '', regex=True)
marvelV0['Budget'] = marvelV0['Budget'].str.replace('\W', '', regex=True)
marvelV0
```

```python
#Sort value related language, set index as title and G
group = marvelV0.sort_values(["Language","Budget"], ascending=False).groupby("Language").head(10)
group.set_index(["Language","Budget"])
marvelV1=group.drop(['Uniq Id', 'Pageurl','Crawl Timestamp','Cast','Collection Date','Filming Locations','Plot'], axis=1)
marvelV1
```

```python
# Analyse correlation beween Budegt, Collection and Review Rating 
columns = ['Budget', 'Collection', 'Review Rating']
subset=marvelV1[columns]
subset.info()
subset['Budget'] = subset['Budget'].astype(float)
subset['Collection'] = subset['Collection'].astype(float)
subset.info()
print(subset.corr())
```

```python
#Set Subset and x, y size 
columns = ['Title','Budget', 'Collection', 'Review Rating']

subset1=marvelV1[columns]
subset1['Budegt/1000000']=subset['Budget']/1000000
subset1['Collection/1000000']=subset['Collection']/1000000
subset1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(x="Budegt/1000000", y="Collection/1000000", data=subset1, hue="Title")
plt.show()
```
